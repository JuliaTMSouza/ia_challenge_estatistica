```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

## Média das vendas

```python
# Transformar dicionario em um Dataframe
df_faturamento_desafio = pd.DataFrame(dict_faturamento)
df_faturamento_desafio.info()
```

```python
df_faturamento_desafio.valor.mean()
```

```python
# Tranformar o campo data_ref para datetime
df_faturamento_desafio.data_ref = pd.to_datetime(df_faturamento_desafio.data_ref)
df_faturamento_desafio.info()
```

```python
ax = df_faturamento_desafio.plot(x='data_ref', y='valor', kind='bar', color='skyblue')
ax.set_xticklabels(df_faturamento_desafio['data_ref'].dt.to_period('M'))
plt.title('Valores por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.show()
```

```python
linha = df_faturamento_desafio.sort_values(by='data_ref').plot.line(x='data_ref', y='valor')
```
